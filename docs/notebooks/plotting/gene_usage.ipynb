```python
import hicutils.core.io as io
import hicutils.core.pooling as pooling
import hicutils.core.filters as filters
import hicutils.plots as plots

df = io.read_directory('../example_data_immunedb')
pdf = (
    df
    .pipe(pooling.pool_by, 'disease') # Pool on disease
    .pipe(filters.filter_functional) # Filter on functional
    .pipe(filters.filter_by_overall_copies, 2) # Filter on 2+ copy
)
```

## Basic VH usage plot

```python
_ = plots.plot_gene_usage(pdf, 'disease', 'v_gene', figsize=(12, 5))
```

## Normalized by column

```python
_ = plots.plot_gene_usage(pdf, 'disease', 'v_gene', normalize_by='cols', figsize=(12, 5))
```

## Weighted by copies

```python
_ = plots.plot_gene_usage(pdf, 'disease', 'v_gene', size_metric='copies', figsize=(12, 5))
```

## Disable clustering

```python
_ = plots.plot_gene_usage(pdf, 'disease', 'v_gene', cluster_by=None, figsize=(12, 5))
```
