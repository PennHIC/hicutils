```python
import hicutils.core.io as io
import hicutils.core.pooling as pooling
import hicutils.core.filters as filters
import hicutils.plots as plots

df = io.read_directory('../example_data_immunedb')
```

## String plot of overlapping clones by replicate

```python
_ = plots.plot_strings(df, 'replicate_name')
```

## String plot of *all* clones by replicate

```python
_ = plots.plot_strings(df, 'replicate_name', only_overlapping=False)
```

## String plot of *all* clones heatmapped by frequency

```python
_ = plots.plot_strings(df, 'replicate_name', only_overlapping=False)
```

## String plot of *all* clones heatmapped by log-scaled frequency

```python
_ = plots.plot_strings(df, 'replicate_name', only_overlapping=False, scale='log')
```

## String plot of top 25 clones, heatmapped by log-scaled frequency, and annotated

```python
_ = plots.plot_strings(df, 'replicate_name', only_overlapping=False, scale=True,
                       limit=25, ylabels='full', figsize=(10, 8))
```

## Track CDR3 AAs through all replicates (across donors)

```python
_ = plots.plot_strings(df, 'subject', only_overlapping=False,
                       overlapping_features=('cdr3_aa',))
```
