```python
import hicutils.core.io as io
import hicutils.core.pooling as pooling
import hicutils.core.filters as filters
import hicutils.plots as plots

df = io.read_directory('../example_data_immunedb')
pdf = (
    df
    .pipe(pooling.pool_by, 'disease') # Pool on disease
    .pipe(filters.filter_functional) # Filter on functional
    .pipe(filters.filter_by_overall_copies, 2) # Filter on 2+ copy
)
```

## Clone size distribution

```python
_ = plots.plot_clone_sizes(pdf, aspect=2)
```

## Clone size distribution with cutoff

```python
_ = plots.plot_clone_sizes(pdf, cutoff=20)
```

## Top clone plots

```python
_ = pdf.groupby('disease').apply(plots.plot_top_clones)
```

## Top clones with custom cutoff

```python
_ = pdf.groupby('disease').apply(plots.plot_top_clones, cutoff=50)
```

## Top clones with custom cutoff & annotation

```python
_ = pdf.groupby('disease').apply(plots.plot_top_clones, cutoff=15,
                                 annotate=['v_gene', 'cdr3_aa'])
```

## Clone size range plot

```python
_ = plots.plot_ranges(pdf, 'disease')
```

## Clone size range plot with custom ranges

```python
_ = plots.plot_ranges(pdf, 'disease', intervals=(10, 100, 1000, 5000))
```
