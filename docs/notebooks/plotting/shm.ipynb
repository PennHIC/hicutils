```python
import hicutils.core.io as io
import hicutils.core.pooling as pooling
import hicutils.core.filters as filters
import hicutils.plots as plots

df = io.read_directory('../example_data_immunedb')
pdf = (
    df
    .pipe(pooling.pool_by, 'disease') # Pool on disease
    .pipe(filters.filter_functional) # Filter on functional
    .pipe(filters.filter_by_overall_copies, 2) # Filter on 2+ copy
)
```

## SHM distribution plot (by *clones*)

```python
_ = plots.plot_shm_distribution(pdf, 'disease', 'clones')
```

## SHM distribution plot (by *copies*)

```python
_ = plots.plot_shm_distribution(pdf, 'disease', 'copies')
```

## Aggregate SHM plot

```python
_ = plots.plot_shm_aggregate(pdf, 'disease')
```

## Aggregate SHM plot as boxplot

```python
_ = plots.plot_shm_aggregate(pdf, 'disease', kind='box')
```

## Basic SHM range plot using boxplot

```python
_ = plots.plot_shm_range(pdf, 'disease')
```

## SHM range plot with custom cutoffs

```python
_ = plots.plot_shm_range(pdf, 'disease', buckets=(1, 2, 5, 10, 15))
```

## SHM range plot with custom colors

```python
import seaborn as sns

_ = plots.plot_shm_range(pdf, 'disease', color=sns.color_palette('Blues'))
```
